```python
import pandas as pd
import requests
from io import StringIO

url = "https://en.wikipedia.org/wiki/List_of_S%26P_500_companies"

headers = {
    "User-Agent": "Mozilla/5.0"
}

response = requests.get(url, headers=headers)
response.raise_for_status()

html = StringIO(response.text)

sp500 = pd.read_html(html)[0]

sp500_tickers = (
    sp500["Symbol"]
    .tolist()
)

pd.DataFrame(sp500_tickers, columns=["Symbol"]).to_csv(
    "sp500_tickers.csv",
    index=False
)

sp500_tickers = [ticker for ticker in sp500_tickers if '.B' not in ticker]

sp500_tickers
```

```python
import yfinance as yf
```

```python
start_date = '2000-01-01'
end_date = '2026-02-17'

historical_prices = yf.download(sp500_tickers, start=start_date, end=end_date )
```

```python
historical_pricesf = historical_prices.loc[:, historical_prices.columns.get_level_values(0) == 'Close']
historical_pricesf.columns = historical_pricesf.columns.droplevel(0)
historical_pricesf.to_csv("sp500_close_prices.csv")
historical_pricesf
```

```python
MIN_REQUIRED_NUM_OBS_PER_TICKET = 100

#Count non missing non values for each ticker
ticker_counts = historical_pricesf.count()
ticker_counts
```

```python
valid_tickers_mask = ticker_counts[ticker_counts >= MIN_REQUIRED_NUM_OBS_PER_TICKET].index
historical_pricesf = historical_pricesf[valid_tickers_mask]
```
